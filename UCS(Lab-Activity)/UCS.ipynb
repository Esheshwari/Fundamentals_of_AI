```python
# STEP 1: Define the Graph (as an Adjacency List with costs)
# Each city is a key. Its value is a list of (neighbour, cost) pairs.
graph = {
    "A": [("B", 1), ("C", 4)],
    "B": [("D", 2), ("E", 5)],
    "C": [("F", 3)],
    "D": [("G", 3)],
    "E": [("G", 1)],
    "F": [("G", 6)],
    "G": [],  # Goal node — no outgoing edges
}
```

```python
# STEP 2: The UCS Function
def ucs(graph, start, goal):
    """
    Performs Uniform Cost Search from start to goal.
    Returns the path and total cost if found, else None.
    """

    # Priority Queue — each entry: (total_cost, current_node, path_taken)
    # heapq always pops the entry with the SMALLEST total_cost first
    priority_queue = []

    # Push the starting node with cost 0 of the form (Queue,cost,node, path)
    heapq.heappush(priority_queue, (0, start, [start]))

    # Visited set — to avoid revisiting nodes
    visited = set()

    # Main Loop
    while priority_queue:
        # Pop the node with the LOWEST cost so far
        cost, node, path = heapq.heappop(priority_queue)

        # Goal Check
        if node == goal:
            return path, cost  # Found! Return path and total cost

        # Skip if already visited (a cheaper path was already processed)
        if node in visited:
            continue

        visited.add(node)  # Mark as visited

        # Explore Neighbours
        for neighbour, edge_cost in graph[node]:
            if neighbour not in visited:
                new_cost = cost + edge_cost  # Accumulate cost
                new_path = path + [neighbour]  # Extend path
                # Every time we push to queue, we save the path history.Uses more memory (copying lists constantly).
                heapq.heappush(priority_queue, (new_cost, neighbour, new_path))

    return None, None  # No path found
```

```python
# STEP 3: Run the Search
start_node = "A"
goal_node = "G"

print("\n Running UCS from", start_node, "to", goal_node)

result_path, total_cost = ucs(graph, start_node, goal_node)

if result_path:
    print(f" Path Found : {result_path}")
    print(f" Total Cost : {total_cost}")
else:
    print("No path found!")
```
